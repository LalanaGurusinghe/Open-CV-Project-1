## 1.2 How to Load a Video File

````python
import cv2  
# Import OpenCV library. This gives access to video, image, and camera functions.


# Create a VideoCapture object and load the video file
source = cv2.VideoCapture(
    r'C:\Users\LalanaGurusinghe\OneDrive - Digital400 PVT LTD\Documents\AI\DAY 03\codes\Samples\road_drive.avi'
)
# VideoCapture creates an object that connects to:
# - a video file
# - or webcam (if we pass 0)
# This object stores video state (current frame position, fps, etc.)


paused = False
# This is a boolean flag.
# It controls whether video frames should be read or not.
# False = video is playing
# True = video is paused


while True:
    # Infinite loop → keeps running until we break manually


    if not paused:
        # Only read new frames when video is NOT paused
        
        ret, img = source.read()
        # source.read() reads ONE frame from the video
        # ret = True if frame successfully read
        # ret = False if video ended or error occurred
        # img = the actual frame (image as NumPy array)

        print(ret)
        # Print True/False just to monitor video status

        if ret == False:
            # If no frame is returned → video finished
            break
            # Break the loop completely


    cv2.imshow('Live Video', img)
    # Display the current frame in a window named "Live Video"
    # If paused == True, the same frame will remain on screen


    key = cv2.waitKey(30) 
    # waitKey waits 30 milliseconds for a key press
    # 30 ms ≈ ~33 FPS playback speed
    # & 0xFF ensures we get correct ASCII value on all systems


    if key == 27:
        # 27 is ASCII value for ESC key
        break
        # Exit the entire loop → close video


    elif key == ord('p'):
        # ord('p') gives ASCII value of 'p'
        paused = True
        # Set pause flag to True
        # This stops source.read() from executing
        print("Paused")


    elif key == ord('q'):
        # ord('q') gives ASCII value of 'q'
        paused = False
        # Set pause flag back to False
        # Video continues from same frame position
        print("Resumed")


source.release()
# Release the video file or camera from memory
# Very important to free system resources


cv2.destroyAllWindows()
# Close all OpenCV display windows
````

# 📌 Point Operations in Image Processing

## 🔹 Introduction

Point operations (also called **pixel-wise operations**) are image processing techniques where:

> Each output pixel value depends only on the corresponding input pixel value.

There is **no relationship with neighboring pixels**.

Mathematically:

\[
g(x, y) = T(f(x, y))
\]

Where:

- `f(x, y)` → Input image pixel value  
- `g(x, y)` → Output image pixel value  
- `T` → Transformation function  

---

## 🔹 Why Point Operations Are Important

- Image enhancement  
- Contrast adjustment  
- Brightness correction  
- Image thresholding  
- Image negative creation  

---

# 📌 Types of Point Operations

---

## 1️⃣ Image Negative

Transforms dark areas to bright and bright areas to dark.

### Formula:

$$
g(x, y) = 255 - f(x, y)
$$

### Example in Python:
```python
import cv2

img = cv2.imread("image.jpg", 0)  # grayscale
negative = 255 - img

cv2.imshow("Negative Image", negative)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

## 2️⃣ Brightness Adjustment

Increase or decrease brightness by adding a constant value to each pixel.

### Formula:

$$
g(x, y) = f(x, y) + C
$$

Where `C` is a constant.

### Example in Python:
```python
bright = cv2.add(img, 50)  # increase brightness by 50
cv2.imshow("Brightness Increased", bright)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

## 3️⃣ Contrast Adjustment (Scaling)

Multiply pixel values by a constant to adjust contrast.

### Formula:

$$
g(x, y) = C \cdot f(x, y)
$$

- `C > 1` → contrast increases  
- `0 < C < 1` → contrast decreases

### Example in Python:
```python
contrast = cv2.convertScaleAbs(img, alpha=1.5, beta=0)  # alpha=contrast
cv2.imshow("Contrast Adjusted", contrast)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

## 4️⃣ Thresholding

Convert grayscale image into a binary image by comparing each pixel with a threshold value.

### Formula:

$$
g(x, y) = \begin{cases} 255 & \text{if } f(x,y) > T \\ 0 & \text{otherwise} \end{cases}
$$

Where `T` is the threshold.

### Example in Python:
```python
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
cv2.imshow("Thresholded Image", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

## 5️⃣ Log Transformation

Used to enhance dark regions while compressing bright regions.

### Formula:

$$
g(x, y) = C \log(1 + f(x, y))
$$

Where `C` is a scaling constant.

### Example in Python:
```python
import numpy as np

c = 255 / np.log(1 + np.max(img))
log_image = c * (np.log(1 + img))
log_image = np.array(log_image, dtype=np.uint8)

cv2.imshow("Log Transformed", log_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

## 6️⃣ Power Law (Gamma) Transformation

Used for gamma correction to adjust image brightness non-linearly.

### Formula:

$$
g(x, y) = C \cdot f(x, y)^\gamma
$$

- `γ < 1` → brighter image  
- `γ > 1` → darker image

### Example in Python:
```python
gamma = 0.5
gamma_corrected = cv2.pow(img / 255.0, gamma) * 255
gamma_corrected = np.array(gamma_corrected, dtype=np.uint8)

cv2.imshow("Gamma Corrected", gamma_corrected)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## THRESH_TOZERO and THRESH_TRUNC

---

### 1️⃣ THRESH_TRUNC (Truncate)

Think of it like a **ceiling** — pixel values are not allowed to go above T.

$$
g(x, y) = \begin{cases} T & \text{if } f(x,y) > T \\ f(x,y) & \text{otherwise} \end{cases}
$$

- Pixels **brighter than T** → get **capped at T**
- Pixels **darker than T** → stay **unchanged**

So the image is never fully black or white — it just **limits the brightness**.

**Example:** If T = 127
- Pixel value 200 → becomes 127
- Pixel value 80 → stays 80
- Pixel value 50 → stays 50
```python
import cv2

img = cv2.imread('image.jpg', 0)

ret, thresh_trunc = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)

cv2.imshow('Original', img)
cv2.imshow('THRESH_TRUNC', thresh_trunc)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

### 2️⃣ THRESH_TOZERO

Think of it like a **filter** — only bright pixels survive, dark ones are wiped out.

$$
g(x, y) = \begin{cases} f(x,y) & \text{if } f(x,y) > T \\ 0 & \text{otherwise} \end{cases}
$$

- Pixels **brighter than T** → stay **unchanged**
- Pixels **darker than T** → become **0 (black)**

**Example:** If T = 127
- Pixel value 200 → stays 200
- Pixel value 80 → becomes 0
- Pixel value 50 → becomes 0
```python
import cv2

img = cv2.imread('image.jpg', 0)

ret, thresh_tozero = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)

cv2.imshow('Original', img)
cv2.imshow('THRESH_TOZERO', thresh_tozero)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

### Quick Comparison Table

| Pixel Value | THRESH_TRUNC (T=127) | THRESH_TOZERO (T=127) |
|-------------|----------------------|-----------------------|
| 200 | 127 (capped) | 200 (unchanged) |
| 150 | 127 (capped) | 150 (unchanged) |
| 80 | 80 (unchanged) | 0 (wiped) |
| 50 | 50 (unchanged) | 0 (wiped) |

---

### Key Difference in One Line:
- **THRESH_TRUNC** → *"Don't let pixels get too bright"*
- **THRESH_TOZERO** → *"Kill dark pixels, keep bright ones as they are"*

### HSV color space

HSV is a cylindrical color model that remaps the RGB primary colors into dimensions that are easier for humans to understand. Like the Munsell Color System, these dimensions are hue, saturation, and value.

Hue specifies the angle of the color on the RGB color circle. A 0° hue results in red, 120° results in green, and 240° results in blue.
Saturation controls the amount of color used. A color with 100% saturation will be the purest color possible, while 0% saturation yields grayscale.
Value controls the brightness of the color. A color with 0% brightness is pure black while a color with 100% brightness has no black mixed into the color. Because this dimension is often referred to as brightness, the HSV color model is sometimes called HSB, including in P5.js.
It is important to note that the three dimensions of the HSV color model are interdependent. If the value dimension of a color is set to 0%, the amount of hue and saturation does not matter as the color will be black. Likewise, if the saturation of a color is set to 0%, the hue does not matter as there is no color used. Because the hue dimension is circular, the HSV color model is best depicted as a cylinder. This is illustrated in the interactive example below, where all possible color mixes are represented within the bounds of the cylinder.


````python
import cv2
import numpy as np

frame=cv2.imread(r'C:\Users\LalanaGurusinghe\OneDrive - Digital400 PVT LTD\Documents\AI\DAY 03\codes\Samples\colors3.jpg')

height,width=frame.shape[0:2]
frame=cv2.resize(frame,(int(height/2),int(width/2)))

hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

lower_color = np.array((0,80,80))
upper_color = np.array((10,255,255))

#Green (59,189,0) (100,255,255)
#Red = (1,136,63), (10,255,255)
#Blue = (96,227,130), (140,255,255)

mask = cv2.inRange(hsv, lower_color, upper_color)

res = cv2.bitwise_and(frame,frame, mask= mask)

cv2.imshow('frame',frame)
cv2.imshow('mask',mask)
cv2.imshow('res',res)
cv2.imshow('hsv',hsv)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Color Detection using HSV Masking in OpenCV

---
```python
import cv2
import numpy as np

# ── STEP 1: Load the image ──────────────────────────────────────────
frame = cv2.imread(r'C:\Users\LalanaGurusinghe\OneDrive - Digital400 PVT LTD\Documents\AI\DAY 03\codes\Samples\colors3.jpg')
# cv2.imread() loads the image in BGR format (Blue, Green, Red)
# This is OpenCV's default color order — NOT RGB like most other libraries

# ── STEP 2: Convert BGR → HSV ───────────────────────────────────────
hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
# HSV stands for:
#   H (Hue)        → the actual COLOR (0–179 in OpenCV)
#   S (Saturation) → how vivid/pure the color is (0–255)
#   V (Value)      → brightness of the color (0–255)
#
# Why HSV instead of BGR?
# Because detecting colors by range is MUCH easier in HSV.
# For example, "blue" in BGR can have hundreds of combinations,
# but in HSV, blue always falls within a small Hue range.

# ── STEP 3: Define the color range to detect ────────────────────────
lower_color = np.array((96, 227, 130))
upper_color = np.array((140, 255, 255))
#
# We are targeting BLUE color here.
# Each value corresponds to (H, S, V):
#
#   lower_color = (96,  227, 130)
#                  H    S    V
#   upper_color = (140, 255, 255)
#                  H    S    V
#
# H range 96–140 → covers blue shades in OpenCV's HSV scale
# S range 227–255 → only vivid/saturated blues (ignores dull/gray)
# V range 130–255 → only reasonably bright blues (ignores very dark)

# ── STEP 4: Create the mask ─────────────────────────────────────────
mask = cv2.inRange(hsv, lower_color, upper_color)
# cv2.inRange() scans every pixel in the HSV image and checks:
#   → If the pixel falls within [lower_color, upper_color] → 255 (WHITE)
#   → If the pixel is outside the range              → 0   (BLACK)
#
# The result is a BINARY IMAGE (mask):
#   White pixels = detected color (blue)
#   Black pixels = everything else

# ── STEP 5: Display the results ─────────────────────────────────────
cv2.imshow('mask', mask)
# Shows the binary mask — white areas are where blue was detected

cv2.imshow('hsv', hsv)
# Shows the HSV version of the image
# Colors will look different from the original — this is normal

cv2.imshow('frame', frame)
# Shows the original BGR image for comparison

# ── STEP 6: Wait and close ──────────────────────────────────────────
cv2.waitKey(0)
# Waits indefinitely until any key is pressed

cv2.destroyAllWindows()
# Closes all OpenCV windows cleanly
```

## Contour Operations in Image Processing

---

### What is a Contour?

A **contour** is a curve that joins all the continuous points along a boundary that have the **same color or intensity**.

Simply put — contours are the **outlines or edges of objects** in an image.

Think of it like tracing the shape of an object with a pen:
```
Original Image          Contour Detected
┌─────────────┐         ┌─────────────┐
│             │         │   ╔═════╗   │
│   ███████   │   →     │   ║     ║   │
│   ███████   │         │   ║     ║   │
│   ███████   │         │   ╚═════╝   │
│             │         │             │
└─────────────┘         └─────────────┘
     Object                 Outline only
```

---

### How Contours Work (Step by Step)
```
Original Image
      ↓
Convert to Grayscale
      ↓
Apply Threshold / Edge Detection
      ↓
Find Contours → cv2.findContours()
      ↓
Draw Contours → cv2.drawContours()
```

---

### Basic Code Example
```python
import cv2
import numpy as np

# ── STEP 1: Load and convert to grayscale ──────────────────────────
img = cv2.imread('image.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Contours work on grayscale or binary images, not color images

# ── STEP 2: Apply threshold to get binary image ────────────────────
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
# Converts image to pure black and white
# This makes it easier for OpenCV to find the edges

# ── STEP 3: Find contours ──────────────────────────────────────────
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# contours  → list of all detected contours (each contour is an array of points)
# hierarchy → relationship between contours (parent/child nesting)

# ── STEP 4: Draw contours ──────────────────────────────────────────
cv2.drawContours(img, contours, -1, (0, 255, 0), 2)
# img        → image to draw on
# contours   → list of contours to draw
# -1         → draw ALL contours (-1 means all, 0,1,2.. draws specific one)
# (0,255,0)  → color in BGR (green in this case)
# 2          → thickness of the contour line

cv2.imshow('Contours', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

### cv2.findContours() Parameters Explained
```python
contours, hierarchy = cv2.findContours(image, mode, method)
```

#### Retrieval Mode (2nd parameter):

| Mode | Description |
|------|-------------|
| `cv2.RETR_EXTERNAL` | Finds only the **outermost** contours |
| `cv2.RETR_INTERNAL` | Finds only **inner/hole** contours |
| `cv2.RETR_TREE` | Finds **all** contours and builds full hierarchy |
| `cv2.RETR_LIST` | Finds all contours but **no hierarchy** |

#### Approximation Method (3rd parameter):

| Method | Description |
|--------|-------------|
| `cv2.CHAIN_APPROX_NONE` | Stores **every single point** on the contour |
| `cv2.CHAIN_APPROX_SIMPLE` | Stores only **corner points** (saves memory) |
```
CHAIN_APPROX_NONE          CHAIN_APPROX_SIMPLE
● ● ● ● ● ● ●              ●           ●
●           ●              
●           ●              
●           ●              ●           ●
● ● ● ● ● ● ●              ●           ●
Stores all points          Stores only corners
```

---

### Useful Contour Features

Once contours are found, you can extract useful information:
```python
for cnt in contours:

    # ── Area ───────────────────────────────────────────────────────
    area = cv2.contourArea(cnt)
    # Returns the area (in pixels) inside the contour
    print("Area:", area)

    # ── Perimeter ──────────────────────────────────────────────────
    perimeter = cv2.arcLength(cnt, True)
    # True = closed contour (shape is closed)
    # False = open contour (curve/line)
    print("Perimeter:", perimeter)

    # ── Bounding Rectangle ─────────────────────────────────────────
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    # Draws a straight rectangle around the contour

    # ── Centroid (center point) ────────────────────────────────────
    M = cv2.moments(cnt)
    if M['m00'] != 0:
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        cv2.circle(img, (cx, cy), 5, (0, 0, 255), -1)
    # Draws a red dot at the center of the object
```

---

### Filtering Contours by Size

Very useful to **ignore noise** and detect only objects of a certain size:
```python
for cnt in contours:
    area = cv2.contourArea(cnt)

    if area > 500:  # ignore contours smaller than 500 pixels
        cv2.drawContours(img, [cnt], -1, (0, 255, 0), 2)
```

````python
import numpy as np
import cv2

img = cv2.imread(r'C:\Users\LalanaGurusinghe\OneDrive - Digital400 PVT LTD\Documents\AI\DAY 03\codes\Samples\gray_image.png')
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 80, 255, 0)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:

    area=cv2.contourArea(cnt)
    if(area>2000 and area<8000.0):
        cv2.drawContours(img, [cnt], -1, (0,255,0), 2) #-1 draw all the contours , green color , line tickness is 3
        x,y,w,h = cv2.boundingRect(cnt)
        print(x,y,w,h)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
        print('Coin Diameter :',h,' Coin Area',area)
    
        
    

#print(len(contours))

#print(contours[6])




cv2.imshow('thresh',thresh)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
